```python
import numpy as np
import soundfile as sf
import librosa
from librosa import display
import matplotlib.pyplot as plt
import sys
from scipy.fft import fft
from scipy import signal
from numpy import savetxt
from numpy import save
import scipy.signal as sg

durations = []
ds_samples = []
ds_fft = []
ds_spectro = []
ds_low =[]
ds_mfccs = []
labels = []

#read audio path references from PATH file
with open("PATH.txt","r") as f:
    count = 0
    for el in f:
        if ".txt" not in el and "LENGTH" not in el:
            file_path = el.split()[0]
            samples, sampling_rate = librosa.load(file_path,
                                                    sr = 16000,
                                                    mono = True,
                                                    offset = 0.0,
                                                    duration = 6.0)
            duration_of_sound = samples.shape[0]/sampling_rate
            
            #read only audio with duration >= 6 sec
            if duration_of_sound >= 6.0:

                labels.append(el.split()[1])

                #frequence-domain representation
                ftt = fft(samples)

                #low pass filter to 250
                b, a = sg.butter(4, 250. / (16000 / 2.), 'low')
                low = sg.filtfilt(b, a, samples)

                #spectrogramma
                frequencies, times, spectrogram = signal.spectrogram(samples, sampling_rate)

                #MFCCS features
                mfccs = librosa.feature.mfcc(y=samples, sr=16000, n_mfcc=40)

                
                # check if data are correctly extracted
                if count == 0:

                    plt.figure()
                    librosa.display.waveplot(y = samples, sr = sampling_rate)
                    plt.xlabel("Time ( seconds ) --> ")
                    plt.ylabel("Amplitude")
                    plt.show()

                    plt.figure()                    
                    length = np.array(ftt).shape[0]/2
                    y = np.arange(length)
                    x1 = np.array_split(ftt, 2)
                    plt.xlabel("Frequences --> ")
                    plt.ylabel("Amplitude")
                    plt.plot(y, x1[0])
                    plt.show()

                    plt.pcolormesh(times, frequencies, spectrogram)
                    plt.imshow(spectrogram)
                    plt.ylabel('Frequency [Hz]')
                    plt.xlabel('Time [sec]')
                    plt.show()
                    print(spectrogram.shape)

                count += 1
                
                # create dataset
                ds_samples.append(samples)
                ds_fft.append(ftt)
                ds_low.append(low)
                ds_spectro.append(spectrogram.flatten())
                ds_mfccs.append(mfccs)

# cast list to array  
ds_samples = np.array(ds_samples)
ds_fft = np.array(ds_fft)
ds_spectro = np.array(ds_spectro)
ds_low = np.array(ds_low)
ds_mfccs = np.array(ds_mfccs)
labels = np.array(labels)

# check on dataset dimentions
print(ds_samples.shape)
print(ds_fft.shape)
print(ds_spectro.shape)
print(ds_low.shape)
print(ds_mfccs.shape)
print(labels.shape)

# save dataset
save('ds_samples.npy', ds_samples)
save('ds_fft.npy', ds_fft)
save('ds_spectro.npy', ds_spectro)
save('ds_low.npy', ds_low)
save('ds_mfccs.npy', ds_mfccs)
save('labels.npy', labels)

f.close()
```
